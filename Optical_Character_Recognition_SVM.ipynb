```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
# reading the CSV file into pandas dataframe
letter_df=pd.read_csv('letterdata.csv')
```

```python
#check count of unique records in 'letter' column
letter_df['letter'].value_counts()
```

```python
# Check top few records
letter_df.head()
```

```python
x,y=np.array(letter_df)[:,1:16],np.array(letter_df.letter)[:] # dependent and independnt variables from the data set
```

```python
x.shape
```

```python
y.shape
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)#split into test and train sets
```

```python
#RBF SVM with optimized hyperparameters 'gamma' and 'C' So that misclassification rate is lower and also not tends to overfit.
clf=svm.SVC(gamma=0.025,C=10)#here gamma='Distance of influence of a single training point',C='Penalty parameter of the error term'
```

```python
#fit the trainning data
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)#accuracy
```

```python
y_pred=clf.predict(x_test)#prediction on test data
pd.crosstab(y_test,y_pred)#cross tabs to show out of total collected record how many are predicted correctly, that are shown here diagonally. 
```
